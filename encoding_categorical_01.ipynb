### To handling infrequent categories

```python
X = np.array([['dog'] * 5 + ['cat'] * 20 + ['rabbit'] * 10 + ['snake'] * 3 + ['horse'] * 2], dtype=object).T
```

```python
X
```

1st approch

```python
enc = OrdinalEncoder(max_categories=3).fit(X)
# it will make only 3 categories and short them according to there values(best example in copy)
```

To see the infrequent categories

```python
enc.infrequent_categories_
```

```python
enc.transform(np.array([['cat', 'rabbit', 'snake', 'dog', 'horse']]).reshape(5,1))
```

Better way <br>
2nd approch

```python
enc = OrdinalEncoder(min_frequency=4).fit(X)
enc.infrequent_categories_
```

```python
enc.transform(np.array([['cat', 'rabbit', 'dog','snake', 'horse']]).reshape(5,1))
```

### 4. LabelBinarizer

```python
from sklearn.preprocessing import LabelBinarizer

# Sample target variables for a multi-class classsification problem
y = ['cat', 'dog', 'fish', 'dog', 'cat']

# Initialize the LabelBinarizer
lb = LabelBinarizer()

# Fit and transform the target variables
y_binarized = lb.fit_transform(y)

print('Binarized Labels:\n', y_binarized)

# Inverse transform to recover Original Labels
y_original = lb.inverse_transform(y_binarized)

print('Original Labels:\n', y_original)
```

###  5. MultiLabelBinarizer

```python
from sklearn.preprocessing import MultiLabelBinarizer

# example multi-label data
y = [('red', 'blue'), ('blue', 'green'), ('green',), ('red',)]

# Initialize MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# fit and transform the data to Binary matrix
y_mlb = mlb.fit_transform(y)

print('Binary matrix:\n', y_mlb)
print('Class labels:', mlb.classes_)

# Inverse transform to recover original labels 
y_inv = mlb.inverse_transform(y_mlb)
print('Inverse transform labels:\n', y_inv)
```
